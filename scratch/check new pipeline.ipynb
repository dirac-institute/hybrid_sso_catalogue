```python
import sys
sys.path.append("../hybridcat/")

import hybridcat
import matplotlib.pyplot as plt
import pandas as pd
```

```python
the_creator = hybridcat.pipeline.HybridCreator(catalogue_folder="../catalogues/", output_folder='../output/',
                                               verbose=True, recreate=True)
```

```python
the_creator.mpcorb = pd.read_hdf(the_creator.catalogue_folder + "mpcorb_propagated_cart.h5", key="df")
the_creator.s3m = pd.read_hdf(the_creator.catalogue_folder + "s3m_propagated_cart.h5", key="df")
```

```python
the_creator.s3m = hybridcat.transform.transform_catalogue(the_creator.s3m, current_coords="CART", transformed_coords="COM",
                                                          initialise=the_creator.init)
```
